```python
import pandas as pd
```

```python
df=pd.read_csv('lastECommerce.csv')
```

```python
df['Comment_id']=df['comment_ids'].apply(lambda x:x.split('-')[-1])
```

```python
top_sell_items_df = df.groupby('product_name').agg({'product_ratings(out of 5)':'sum'})

top_sell_items_df.reset_index(inplace=True)


    # Rank the product by most Qty sold
top_sell_items_df['Top_Sell_Rank'] = top_sell_items_df['product_ratings(out of 5)'].rank(method='min',ascending=False).astype(int)
```

```python
top_sell_items_df.sort_values(by='Top_Sell_Rank',ascending=True)
```
