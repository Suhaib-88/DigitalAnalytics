```python
import pandas as pd
import glob
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('FB_Analytics.csv')
```

```python
df['Page total likes']=df['Page total likes'].apply(lambda x:x//4)
```

```python
plt.figure(figsize=(6,5))
df.groupby('Post Month')['Page total likes'].mean().plot(kind='bar')
plt.ylabel('Likes count')
plt.savefig('Monthlylikes_fb5.png')
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(df['Post Month'],df['Page total likes'])
plt.ylabel('Page Likes')
plt.savefig('Monthlylikes_Box5.png')
```

```python
df.drop(df[df.like>350].index,axis=0,inplace=True)
```

```python
df=df.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(10,8))
with sns.axes_style('white'):
    g = sns.pointplot(x='Category', y='comment', data=df,color='coral',hue='Type',ax=ax[0]);
    g = sns.pointplot(x='Category', y='like', data=df,color='yellow',hue='Type',ax=ax[1]);
    g = sns.pointplot(x='Category', y='share', data=df,color='green',hue='Type',ax=ax[2]);
plt.xlabel('category')
plt.ylabel('shares')
plt.legend()
fig.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.6)
fig.suptitle("Users category Vs comment,likes,shares", fontsize=14)
plt.savefig('likevscommnets5.png')
```

### Paid vs unpaid

```python
print('unpaid Lifetime Post Impressions by people who have liked your Page :',df[df['Paid']==0]['Lifetime Post Impressions by people who have liked your Page'].mean()/(df[df['Paid']==0]['Lifetime Post Impressions by people who have liked your Page'].mean() + df[df['Paid']==1]['Lifetime Post Impressions by people who have liked your Page'].mean()))
print('unpaid Lifetime Post reach by people who like your Page :',df[df['Paid']==0]['Lifetime Post reach by people who like your Page'].mean()/(df[df['Paid']==0]['Lifetime Post reach by people who like your Page'].mean() + df[df['Paid']==1]['Lifetime Post reach by people who like your Page'].mean()))
print('unpaid Lifetime People who have liked your Page and engaged with your post :',df[df['Paid']==0]['Lifetime People who have liked your Page and engaged with your post'].mean()/(df[df['Paid']==0]['Lifetime People who have liked your Page and engaged with your post'].mean() + df[df['Paid']==1]['Lifetime People who have liked your Page and engaged with your post'].mean()))
```
