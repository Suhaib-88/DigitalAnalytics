```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('InstagAnalytics.csv')
```

```python
df2=pd.read_csv('comments (2).csv')
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['months']=df.Date.apply(lambda x:x.month)
df['years']=df.Date.apply(lambda x:x.year)
df['days']=df.Date.apply(lambda x:x.day)
```

```python
plt.figure(figsize=(6,5))
df.groupby('months')['number_of_Likes'].sum().plot(kind='bar')
plt.ylabel('monthly likes')
plt.savefig('monthly_likes.jpg')
```

```python
plt.figure(figsize=(6,5))
df.groupby('months')['number_of_Comments'].sum().plot(kind='bar')
plt.ylabel('monthly comments')
plt.savefig('monthlyComments.jpg')
```

```python
plt.figure(figsize=(6,5))
df.groupby('years')['number_of_Comments'].sum().plot(kind='bar')
plt.ylabel('Yearly comments')
plt.savefig('yearlyComments.jpg')
```

```python
plt.figure(figsize=(6,5))
df.groupby('years')['number_of_Likes'].sum().plot(kind='bar')
plt.ylabel('Yearly likes')

plt.savefig('yearlyLikes.jpg')
```

```python
plt.figure(figsize=(6,5))
df.groupby('days')['number_of_Likes'].sum().plot(kind='bar')
plt.ylabel('daily likes')

plt.savefig('dailyLikes.jpg')
```

```python
plt.figure(figsize=(6,5))
df.groupby('days')['number_of_Comments'].sum().plot(kind='bar')
plt.ylabel('daily comments')

plt.savefig('dailyComments.jpg')
```
